## Introducción:  
La National Futbol League o NFL, al final de cada temporada realiza el prestigioso Draft, el cual tiene como objetivo elegir a los jugadores más valiosos de la NCAA para darles la oportunidad de jugar en la liga. De los 32 equipos de la NFL combinados eligen un total de 259 jugadores de cualquier parte del mundo, claro con algúnas excepciones que no nos interesan. El objetivo del draft es elegir a los jugadores más destacados para ayudar a su equipo a subir en la tabla o incluso llegar a ganar un Super Bowl, a pesar de esto existen muchos casos en los que los jugadores más populares como campeones de la NCAA o favoritos del público, suelen no cumplir con los estándares de rendimiento que los analistas esperan de ellos.

## Antecedentes

### NFL Draft:  

1. Selección de Jugadores:
El momento más crucial del draft es cuando cada equipo hace su selección. Esto determina qué jugador se unirá al equipo y en qué posición.
2. Posición de Selección:
La posición en la que un equipo selecciona en cada ronda es crítica. Los equipos más débiles obtienen selecciones más tempranas, lo que les da la oportunidad de elegir a los jugadores más talentosos disponibles.
3. Elección Estratégica:
Los equipos deben tomar decisiones estratégicas sobre qué posición y qué jugador elegir. Esto depende de las necesidades del equipo, la disponibilidad de talento y la estrategia a largo plazo.
4. Impacto en el Equipo:
Las selecciones en el draft pueden tener un impacto inmediato en el desempeño del equipo. Un jugador estrella puede cambiar la dinámica de un equipo en la próxima temporada.
5. Desarrollo a Largo Plazo:
Las selecciones también tienen un impacto a largo plazo en el éxito del equipo. Jugadores bien seleccionados y bien desarrollados pueden convertirse en pilares fundamentales durante muchos años.
6. Construcción del Equipo:
El draft es una parte clave de la construcción de un equipo ganador a largo plazo. Las selecciones sabias pueden marcar la diferencia en la competitividad del equipo durante varios años.
7. Oportunidades para Jugadores:
Para los jugadores, ser seleccionado en el draft es una gran oportunidad. Les da la posibilidad de entrar en la NFL y demostrar su talento a nivel profesional.
8. Expectativas de Aficionados:
Las selecciones del draft a menudo generan muchas expectativas entre los aficionados. Los aficionados están ansiosos por ver a los nuevos jugadores en acción y tienen grandes esperanzas en su contribución al equipo.
9. Evaluación Post-Draft:
Después del draft, los expertos y aficionados evalúan las selecciones de cada equipo. Esto puede determinar la percepción sobre qué equipos hicieron un buen trabajo y cuáles podrían haber tomado decisiones más acertadas.
10. Legado del Draft:
A lo largo del tiempo, el éxito o el fracaso de un draft puede influir en la historia y el legado de un equipo.

### NFL COMBINE  
El Combine proporciona a los equipos datos y evaluaciones estandarizadas.

1. Evaluación Integral de Jugadores:

- Medir Habilidades Físicas: Las pruebas físicas, como la velocidad, fuerza, agilidad y habilidades atléticas, proporcionan una evaluación cuantitativa del potencial atlético de un jugador.

- Evaluar Habilidades Técnicas: Se observa a los jugadores en ejercicios específicos de fútbol americano para evaluar su destreza y capacidad para realizar tareas fundamentales de su posición.

- Examinar Estado de Salud: Los exámenes médicos y pruebas de imagen permiten a los equipos identificar posibles problemas de salud y lesiones.

- Entrevistar a los Jugadores: Los equipos pueden interactuar con los jugadores para evaluar su inteligencia futbolística, habilidades de liderazgo y personalidad.

2. Influencia en las Decisiones del Draft:

- Impactar el Ranking de Jugadores: Un desempeño impresionante puede elevar la posición de un jugador en el ranking de selecciones.

- Destacar Jugadores con Potencial no Reconocido: Algunos jugadores pueden sobresalir en el Combine, lo que puede llevar a un aumento en su atractivo para los equipos.

- Alertar sobre Problemas de Salud: Los exámenes médicos pueden revelar problemas de salud que podrían afectar el rendimiento a largo plazo de un jugador.

3. Facilita la Toma de Decisiones:

- Comparar a los Jugadores: Permite comparar a los jugadores en una serie de métricas objetivas.

- Validar la Información del College: Brinda una oportunidad para confirmar o refutar la información proporcionada por los programas universitarios.

- Tomar Decisiones Informadas: Ayuda a los equipos a tomar decisiones más informadas sobre qué jugadores seleccionar en el draft.

Tras una serie de experimentos con la nueva base de datos mergeada, nos dimos cuenta que nuestro modelo no funciona tan bien, pero si solo usamos información del dataset solo del combine. Asi que ahora vamos a hacele una limpieza a los datos:

```python
# Vamos a remplazar los NaN numéricos por el promedio de la variable.
def reemplazar_nans_con_promedio(dataframe, columnas):
    for columna in columnas:
        promedio = dataframe[columna].mean()
        dataframe[columna].fillna(promedio, inplace=True)

columnas_a_procesar = ['Age', 'Sprint_40yd', 'Vertical_Jump', 'Bench_Press_Reps', 'Broad_Jump', 'Agility_3cone', 'Shuttle']

# Llamada a la función
reemplazar_nans_con_promedio(combine_data, columnas_a_procesar)
```

```python
# Vimos que los NaN en la variable 'Drafted..tm.rnd.yr.' son simplemente porque no se draftearon por lo cuál no tienen esa información.
combine_data['Drafted..tm.rnd.yr.'].fillna('No Drafted', inplace=True)
```

```python
combine_data[['drafted_by', 'round_drafted', 'pick_drafted', 'year_drafted']] = combine_data['Drafted..tm.rnd.yr.'].str.split('/', expand=True)

# Elimina los espacios en blanco alrededor de cada valor
combine_data = combine_data.apply(lambda x: x.str.strip() if x.dtype == 'object' else x)

combine_data['round_drafted'].fillna('No Drafted', inplace=True)
combine_data['pick_drafted'].fillna('No Drafted', inplace=True)
combine_data['year_drafted'].fillna('No Drafted', inplace=True)
combine_data = combine_data.drop('Player', axis=1)
combine_data = combine_data.drop('Drafted..tm.rnd.yr.', axis=1)

combine_data.isna().sum()
```

```python
import mlflow
from mlflow.models import infer_signature

from sklearn.model_selection import train_test_split
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor

#Define el nombre de tu experimento
experiment_name = "draft1"

experiment_id = mlflow.create_experiment(experiment_name)

print(experiment_id)
```

```python
mlflow.sklearn.autolog()
with mlflow.start_run(experiment_id=experiment_id, run_name="Random Forest") as run:
    # Load the diabetes dataset.
    db = load_diabetes()
    X_train, X_test, y_train, y_test = train_test_split(db.data, db.target)

    # Create and train models.
    rf = RandomForestRegressor(n_estimators=100, max_depth=6, max_features=3)
    rf.fit(X_train, y_train)

    # Use the model to make predictions on the test dataset.
    predictions = rf.predict(X_test)
    print(predictions)

    signature = infer_signature(X_test, predictions)
    mlflow.sklearn.log_model(rf, "model", signature=signature)
```
